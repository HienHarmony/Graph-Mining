### Yêu cầu 2:
- Sinh ngẫu nhiên một ma trận kích thước $10 \times 10$ để làm ma trận liên thuộc của một graph có hướng $10$ đỉnh.
- Tạo graph có hướng $H$ từ ma trận trên.
- Thực hiện vẽ graph $H$ (có mũi tên) theo đoạn code bên dưới:

```python

import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

A = np.random.randint(0, 2, size=(10, 10))

H = nx.from_numpy_array(A, create_using=nx.DiGraph)

```

```python
options = {
    'node_color': 'blue',
    'node_size': 100,
    'width': 3,
    'arrowstyle': '-|>',
    'arrowsize': 12,
}
nx.draw_networkx(H, arrows=True, with_labels=True, **options)
```

- Tham khảo link sau đây để chuyển graph trên sang dạng Json, thử đọc lại graph từ file Json lưu lại được::

https://networkx.org/documentation/stable/reference/readwrite/json_graph.html

- Tham khảo link sau đây để chuyển graph trên sang dạng GML, thử đọc lại graph từ file GML lưu lại được:

https://networkx.org/documentation/stable/reference/readwrite/gml.html

```python

import json
import networkx as nx

with open('graph.json', 'w') as f:
    json.dump(nx.node_link_data(H), f)

with open('graph.json') as f:
    data = json.load(f)
    G = nx.node_link_graph(data)

print(G)
```

```python
nx.write_gml(G, "graph.gml")

```

```python
G_read = nx.read_gml("graph.gml")
print(G_read)
```

```python
print(G_read.nodes)
print(G_read.edges)
```

```python
colors = []
S = G.degree(H)
for v in H.nodes:
    if S[v] % 2 == 1:
        colors.append('green')
    else:
        colors.append('red')

nx.draw(H, with_labels=True, node_color=colors)
plt.show()
```

### Yêu cầu 3:
- Viết hàm sinh ngẫu nhiên một graph với $n$ đỉnh và desity là $p$. Từ đó so sánh với hàm erdos_renyi_graph và gnm_random_graph có sẵn của thư viện.
- Vẽ tất cả các graph sau đây bằng cách tạo và vẽ graph ứng với thư viện có sẵn: cycle_graph, path_graph, star_graph, complete_graph, wheel_graph, full_rary_tree, grid_2d_graph, ego_graph, windmill_graph.

```python
import random
def random_graph(n, p):

    G = nx.Graph()
    for i in range(n):
        G.add_node(i)
    for i in range(n):
        for j in range(i + 1, n):
            if random.random() <= p:
                G.add_edge(i, j)

    return G


n = 100
p = 0.5

G_random = random_graph(n, p)
G_erdos_renyi = nx.erdos_renyi_graph(n, p)
G_gnm = nx.gnm_random_graph(n, int(n * p))

print(f"Number of nodes in G_random: {G_random.number_of_nodes()}")
print(f"Number of edges in G_random: {G_random.number_of_edges()}")
print(f"Number of nodes in G_erdos_renyi: {G_erdos_renyi.number_of_nodes()}")
print(f"Number of edges in G_erdos_renyi: {G_erdos_renyi.number_of_edges()}")
print(f"Number of nodes in G_gnm: {G_gnm.number_of_nodes()}")
print(f"Number of edges in G_gnm: {G_gnm.number_of_edges()}")


nx.draw(G_random, with_labels=True)
plt.show()

nx.draw(G_erdos_renyi, with_labels=True)
plt.show()

nx.draw(G_gnm, with_labels=True)
plt.show()
```
